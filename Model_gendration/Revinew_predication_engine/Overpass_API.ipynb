```python
import requests
import json

latitude = 13.0827
longitude = 80.2707
radius = 2000

# Overpass QL query
query = f"""
[out:json];
node
  (around:{radius},{latitude},{longitude})
  [amenity];
out;
"""

url = "http://overpass-api.de/api/interpreter"
response = requests.get(url, params={'data': query})
data = response.json()

# Extract business names and types
businesses = []
for element in data['elements']:
    tags = element.get('tags', {})
    name = tags.get('name', 'Unnamed')
    amenity = tags.get('amenity', 'Unknown')
    businesses.append({"name": name, "type": amenity})

# Print results
print(f"Found {len(businesses)} businesses within {radius/1000} km:")
for b in businesses:  # show first 20
    print(f"- {b['name']} ({b['type']})")

# Save to JSON file
with open("businesses.json", "w", encoding="utf-8") as f:
    json.dump(businesses, f, ensure_ascii=False, indent=2)
```
